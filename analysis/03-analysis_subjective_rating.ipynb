# Load Raw Data

- Exclude pair 0-3 due to the implementation errors
- Exclude pair 7 since their avatar connection was lost midway, and only the voice remained.

```python
folder_path = '../Data'

rating = pd.DataFrame([])
for i in range(4,31):
    if i == 7: continue
    for p in ['a', 'b']:
        pid = f'{i}{p}'
        _rating = pd.read_table(f'{folder_path}/RawData/{pid}_post_conversation.tsv')
        _rating['conversation'] = i
        _rating['side'] = p
        _rating['pid'] = pid
        rating = pd.concat([rating,_rating])
rating
```

# Cronbach Alpha

```python
reliability = rating[['ReliabilityVoice', 'ReliabilityGaze', 'ReliabilityPose', 'ReliabilityFacial']]
res = pg.cronbach_alpha(data=reliability)
print(f"[Reliability] Cronbach's alpha mean is {res[0]:.3}.")
```

```python
attention = rating.filter(regex=r'Attention')
attention.loc[:,"AttentionDistractionSelf"] = 100 - attention["AttentionDistractionSelf"]
attention.loc[:,"AttentionDistractionOther"] = 100 - attention["AttentionDistractionOther"]
attention.loc[:,"AttentionIgnoreSelf"] = 100 - attention["AttentionIgnoreSelf"]
attention.loc[:,"AttentionIgnoreOther"] = 100 - attention["AttentionIgnoreOther"]
res = pg.cronbach_alpha(data=attention)
print(f"[Attention] Cronbach's alpha mean is {res[0]:.3}.")
```

```python
connection = rating.filter(regex=r'Connection')
res = pg.cronbach_alpha(data=connection)
print(f"[Connection] Cronbach's alpha mean is {res[0]:.3}.")
```

```python
colocation = rating.filter(regex=r'Colocation')
colocation.loc[:,"ColocationNotNoticeSelf"] = 100 - colocation["ColocationNotNoticeSelf"]
colocation.loc[:,"ColocationNotNoticeOther"] = 100 - colocation["ColocationNotNoticeOther"]
colocation.loc[:,"ColocationDifferentLocationSelf"] = 100 - colocation["ColocationDifferentLocationSelf"]
colocation.loc[:,"ColocationDifferentLocationOther"] = 100 - colocation["ColocationDifferentLocationOther"]
res = pg.cronbach_alpha(data=colocation)
print(f"[Colocation] Cronbach's alpha mean is {res[0]:.3}.")
```

```python
likability = rating[['Like', 'MeetAgain', 'Attractive']]
res = pg.cronbach_alpha(data=likability)
print(f"[Likability] Cronbach's alpha mean is {res[0]:.3}.")
```

```python
rating.loc[:,'Likability'] = likability.mean(axis=1)
rating.loc[:,'Attention'] = attention.mean(axis=1)
rating.loc[:,'Connection'] = connection.mean(axis=1)
rating.loc[:,'Colocation'] = colocation.mean(axis=1)
```

# Merge for across-participant analysis

```python
_voice = pd.read_csv("processed/voice_across.csv")
_gaze = pd.read_csv("processed/gazes_across.csv")
_nodding = pd.read_csv("processed/nodding_across.csv")
```

```python
merged = pd.merge(_voice, _nodding, on='pid', how='inner')
merged = pd.merge(merged, _gaze, on='pid', how='inner')
merged = pd.merge(merged, rating, on='pid', how='inner')
merged
```

```python
merged.to_csv("processed/subjective_rating.csv")
```
